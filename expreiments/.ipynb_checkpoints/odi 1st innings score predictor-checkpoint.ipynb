```python
inng_1_balls = balls_df[balls_df["inning_no"] == 1]
```

```python
inng_1_balls["current_score"] = inng_1_balls.groupby("match_id")["total_runs"].cumsum()
```

```python
inng_1_balls["final_score"] = inng_1_balls.groupby("match_id")["total_runs"].transform("sum")
```

```python
inng_1_balls["is_wicket"] = np.where(inng_1_balls["wicket_type"] == "retired hurt", 0,
         np.where(inng_1_balls['wicket_type'] == "0" , 0, 1))
```

```python
inng_1_balls["wickets_fallen"] = inng_1_balls.groupby("match_id")["is_wicket"].cumsum()
```

```python
inng_1_balls["wickets_left"] = 10 - inng_1_balls["wickets_fallen"]
```

```python
inng_1_balls["extra_type"].value_counts()
```

```python
inng_1_balls["legal_ball"] = np.where((inng_1_balls["extra_type"] == "noballs") | 
                     (inng_1_balls["extra_type"] == "wides"),
                     0,1)
```

```python
inng_1_balls["legal_ball"].value_counts()
```

```python
inng_1_balls["ball_num_in_over"] = inng_1_balls.groupby(["match_id", "over_num"])["legal_ball"].\
                                    cumsum()
```

```python
inng_1_balls["ball_num_in_over"] = np.where(inng_1_balls["ball_num_in_over"] > 6 ,
                                           np.nan, inng_1_balls["ball_num_in_over"] )
```

```python
inng_1_balls["ball_num_in_over"].fillna(method="bfill", inplace =True)
```

```python
inng_1_balls["current_ball_no"] = (inng_1_balls["over_num"]  * 6) + inng_1_balls["ball_num_in_over"]
```

```python
inng_1_balls["balls_left"] = 300 - inng_1_balls["current_ball_no"]
```

```python
inng_1_balls["run_rate"] = (inng_1_balls["current_score"] * 6) / inng_1_balls["current_ball_no"]
```

```python
female_matches = matches_df[matches_df["gender"] == "female"]["match_id"].to_list()
```

```python
inng_1_balls = inng_1_balls[~inng_1_balls["match_id"].isin(female_matches)]
```

```python
groups = inng_1_balls.groupby(["match_id"])
unique_match_id = inng_1_balls["match_id"].unique()

runs_in_last_30_balls = []
for i in unique_match_id:
    runs_in_last_30_balls.extend(groups.get_group(i).rolling(window = 30).sum()["total_runs"].\
                                values.tolist())
    
```

```python
inng_1_balls["runs_in_last_30_balls"] = runs_in_last_30_balls
```

```python
inng_1_balls["rr_in_last_30_overs"] = inng_1_balls["runs_in_last_30_balls"] / 5
```

```python
inng_1_balls
```

```python
model_df = inng_1_balls[["batting_team", "bowling_team","venue", "current_score", "run_rate", "balls_left",
         "wickets_left","runs_in_last_30_balls","rr_in_last_30_overs", "final_score"]]
```

```python
model_df
```

```python
model_df = model_df.dropna()
```

```python
teams = ["Australia", "Sri Lanka", "India", "England","South Africa", "Pakistan","Bangladesh", "Afghanistan", "Netherlands", "New Zealand"]
venues= ["Arun Jaitley Stadium", "Bharat Ratna Shri Atal Bihari Vajpayee Ekana Cricket Stadium", "Eden Gardens",
         "Himachal Pradesh Cricket Association Stadium", "M Chinnaswamy Stadium", "MA Chidambaram Stadium, Chepauk", "Maharashtra Cricket Association Stadium",
         "Rajiv Gandhi International Stadium, Uppal", "Wankhede Stadium","Narendra Modi Stadium, Ahmedabad"]
```

```python
    def other_team(team):
        return team if team in teams else "other"
    model_df["batting_team"] = model_df["batting_team"].apply(other_team)
    model_df["bowling_team"] = model_df["bowling_team"].apply(other_team)
```

```python
def other_team(team):
    return team if team in venues else "other"
model_df["venue"] = model_df["venue"].apply(other_team)
```

```python
from sklearn.model_selection import train_test_split
```

```python
5*42
```

```python
model_df = pd.get_dummies(model_df)
```

```python
X_train , X_test, y_train , y_test = train_test_split(model_df.drop("final_score", axis = 1),
                                                     model_df["final_score"], test_size=0.2, random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc =StandardScaler()
```

```python
scaled_train = sc.fit_transform(X_train)
scaled_test = sc.transform(X_test)
```

```python
from xgboost import XGBRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.neighbors import KNeighborsRegressor
```

```python
xgb = KNeighborsRegressor()
```

```python
xgb.fit(scaled_train, y_train)
```

```python
xgb.predict(scaled_train)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

```python
print(r2_score(y_test, xgb.predict(scaled_test)))
print(mean_absolute_error(y_test, xgb.predict(scaled_test)))
print(mean_squared_error(y_test, xgb.predict(scaled_test)))
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
xbg_r = XGBRegressor(objective="reg:squarederror")
param_grid = {
    'n_estimators': [100, 200,150],  # Number of boosting rounds (trees)
    'learning_rate': [0.05, 0.2],  # Step size shrinkage
    'max_depth': [3, 5,7],  # Maximum depth of trees
    'subsample': [0.8, 1.0],  # Fraction of samples used for each boosting round
    'colsample_bytree': [0.8, 1.0],
}
```

```python
grid_search = GridSearchCV(
    estimator=xbg_r,
    param_grid=param_grid,
    scoring='accuracy',  # Use the appropriate scoring metric for your problem
    cv=3,  # Number of cross-validation folds
    n_jobs=-1  # Use all available CPU cores
)
```

```python
grid_search.fit(scaled_train, y_train)
```

```python
grid_search.best_params_
```

```python
print(r2_score(y_test, grid_search.best_estimator_.predict(scaled_test)))
```

```python
print(mean_absolute_error(y_test, grid_search.best_estimator_.predict(scaled_test)))
```

```python
print(mean_squared_error(y_test, grid_search.best_estimator_.predict(scaled_test)))
```
